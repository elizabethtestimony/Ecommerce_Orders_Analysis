```python
#Monthly total sales
monthly_sales = df.groupby('Month')['Order_Amount'].sum()
print("Total sales in each month:\n", monthly_sales)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualize monthly total sales
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
plt.title('Monthly Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

AVERAGE SALES IN EACH QUARTER

```python
Quarter_sales = df.groupby('Quarter')['Order_Amount'].mean()
print("Total sales in each Quarter:\n", Quarter_sales)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=Quarter_sales.index, y=Quarter_sales.values, palette='viridis')
plt.title('Quarter Total Sales')
plt.xlabel('Quarter')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
```

ANALYSIS OF THE NUMBER OF CLIENTS IN EACH AGE GROUP

```python
# Define age bins and labels
age_bins = [18, 24, 30, 36, 42, 48, 54, 60, 66, 72, 78, 84]
age_labels = ['18-23', '24-29', '30-35', '36-41', '42-47', '48-53', '54-59', '60-65', '66-71', '72-77', '78-83']

# Create a new column 'Age_Group' based on 'Customer_Age'
df['Age_Group'] = pd.cut(df['Customer_Age'], bins=age_bins, labels=age_labels, right=False)

# Count unique customers per age group
age_group_client_counts = df.groupby('Age_Group')['Customer_ID'].nunique().reset_index()

display(age_group_client_counts)
```

```python
# Visualize the age group client counts
plt.figure(figsize=(14, 7))
sns.barplot(x='Age_Group', y='Customer_ID', data=age_group_client_counts, palette='coolwarm')
plt.title('Number of Unique Clients per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Unique Clients')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

TOP 10 COUNTRIES IN TERMS OF NUMBER OF CUSTOMERS

```python
# Count unique customers per country (displaying top 10)
country_client_counts = df.groupby('Country')['Customer_ID'].nunique().sort_values(ascending=False).reset_index()

print("Top 10 countries by number of unique clients:\n")
display(country_client_counts.head(10))
```

```python
# Visualize the top 10 countries by client counts
plt.figure(figsize=(14, 7))
sns.barplot(x='Country', y='Customer_ID', data=country_client_counts.head(10), palette='crest')
plt.title('Top 10 Countries by Number of Unique Clients')
plt.xlabel('Country')
plt.ylabel('Number of Unique Clients')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

PRODUCT CATEGORY RATINGS ANALYSIS

```python
product_category_ratings = df.groupby('Product_Category')['Review_Rating'].mean().sort_values(ascending=False).reset_index()
print("Average review rating for each product category:\n")
display(product_category_ratings)
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Product_Category', y='Review_Rating', data=product_category_ratings, palette='coolwarm')
plt.title('Average Review Rating per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Review Rating')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

PRODUCT CATEGORY PROFIT ANALYSIS

```python
product_category_profit = df.groupby('Product_Category')['Profit_Amount'].sum().sort_values(ascending=False).reset_index()
print("Total profit amount for each product category:\n")
display(product_category_profit)
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Product_Category', y='Profit_Amount', data=product_category_profit, palette='magma')
plt.title('Total Profit Amount per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit Amount')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

RELATIONSHIP BETWEEN RATINGS AND PROFITS

```python
merged_category_data = pd.merge(product_category_ratings, product_category_profit, on='Product_Category')
display(merged_category_data)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Review_Rating', y='Profit_Amount', data=merged_category_data, hue='Product_Category', s=200)

for i, row in merged_category_data.iterrows():
    plt.text(row['Review_Rating'] + 0.005, row['Profit_Amount'] + 5000, row['Product_Category'],
             horizontalalignment='left', size='small', color='black', weight='semibold')

plt.title('Relationship Between Product Category Ratings and Profit')
plt.xlabel('Average Review Rating')
plt.ylabel('Total Profit Amount')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

RELATIONSHIP BETWEEN CUSTOMER AGE GROUPS AND TOTAL PROFIT

```python
age_group_profit = df.groupby('Age_Group')['Profit_Amount'].sum().sort_values(ascending=False).reset_index()
print("Total profit amount for each age group:\n")
display(age_group_profit)
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Age_Group', y='Profit_Amount', data=age_group_profit, palette='viridis')
plt.title('Total Profit Amount per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Profit Amount')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

MOST ORDERED PRODUCT CATEGORY PER AGE GROUP

```python
# Group by Age_Group and Product_Category to count orders
age_category_orders = df.groupby(['Age_Group', 'Product_Category']).size().reset_index(name='Order_Count')

# Find the product category with the most orders for each age group
most_ordered_category_per_age_group = age_category_orders.loc[age_category_orders.groupby('Age_Group')['Order_Count'].idxmax()]

print("Most ordered product category per age group:\n")
display(most_ordered_category_per_age_group)
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(x='Age_Group', y='Order_Count', hue='Product_Category', data=most_ordered_category_per_age_group, palette='tab10', dodge=False)

plt.title('Most Ordered Product Category per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
